# Import required libraries

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# Load the dataset

```python
iris=load_iris()
```

```python
iris
```

```python
x=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
x
```

```python
y=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
y
```

# Explore the dataset

```python
x.head()
```

```python
x.shape
```

```python
y.shape
```

```python
x.info()
```

```python
pd.get_dummies(y)
```

# Divide data into training and testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Build model now

```python
dtmodel=DecisionTreeClassifier()
```

```python
dtmodel.fit(x_train,y_train)
```

```python
dtmodel
```

```python
dtmodel2=DecisionTreeClassifier(criterion='entropy')
```

```python
dtmodel2.fit(x_train,y_train)
```

# Do the prediction

```python
y_pred=dtmodel.predict(x_test)
```

```python
y_pred
```

# Evlauate the model

```python
species=np.array(y_test).argmax(axis=1)
pred=np.array(y_pred).argmax(axis=1)
confusion_matrix(species,pred)
```

# Plot the decision tree

```python
from matplotlib import pyplot as plt
from sklearn import tree

fig=plt.figure(figsize=(25,30))
tree.plot_tree(dtmodel,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```

```python
from matplotlib import pyplot as plt
from sklearn import tree

fig=plt.figure(figsize=(25,30))
tree.plot_tree(dtmodel2,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```
